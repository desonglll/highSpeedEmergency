```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


def combine_three_images_with_pyplot(img1_path, img2_path, img3_path, orientation='horizontal'):
    # 读取图片
    img1 = mpimg.imread(img1_path)
    img2 = mpimg.imread(img2_path)
    img3 = mpimg.imread(img3_path)

    # 创建图形和子图
    if orientation == 'horizontal':
        fig, ax = plt.subplots(1, 3, figsize=(18, 6))  # 1行3列
        ax[0].imshow(img1)
        ax[0].axis('off')  # 不显示坐标轴
        ax[0].set_title('Monitoring Point 2 (32.31.250.105) Video 1')

        ax[1].imshow(img2)
        ax[1].axis('off')  # 不显示坐标轴
        ax[1].set_title('Monitoring Point 2 (32.31.250.105) Video 2')

        ax[2].imshow(img3)
        ax[2].axis('off')  # 不显示坐标轴
        ax[2].set_title('Monitoring Point 2 (32.31.250.105) Video 3')

    elif orientation == 'vertical':
        fig, ax = plt.subplots(3, 1, figsize=(6, 18))  # 3行1列
        ax[0].imshow(img1)
        ax[0].axis('off')  # 不显示坐标轴
        ax[0].set_title('Image 1')

        ax[1].imshow(img2)
        ax[1].axis('off')  # 不显示坐标轴
        ax[1].set_title('Image 2')

        ax[2].imshow(img3)
        ax[2].axis('off')  # 不显示坐标轴
        ax[2].set_title('Image 3')

    # 调整子图间距
    plt.tight_layout()

    # 显示合并后的图片
    plt.show()
```
