```python
from tools.csv import csv_files
csv_list =csv_files("./res/with-speed-combined")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
  # 替换为你的CSV文件路径
def lrm(path):
    data = pd.read_csv(path)
    
    # 提取数据
    frames = data['Frame']
    flow = data['Flow']
    density = data['Density']
    speed = data['Speed']
    
    # 创建图形和子图
    fig, axs = plt.subplots(3, 1, figsize=(10, 15))
    
    # 绘制流量（Flow）趋势图
    axs[0].plot(frames, flow, color='blue', label='Flow')
    axs[0].set_title('Traffic Flow Over Time')
    axs[0].set_xlabel('Frame')
    axs[0].set_ylabel('Flow')
    axs[0].legend()
    axs[0].grid()
    
    # 绘制密度（Density）趋势图
    axs[1].plot(frames, density, color='green', label='Density')
    axs[1].set_title('Traffic Density Over Time')
    axs[1].set_xlabel('Frame')
    axs[1].set_ylabel('Density')
    axs[1].legend()
    axs[1].grid()
    
    # 绘制速度（Speed）趋势图
    axs[2].plot(frames, speed, color='red', label='Speed')
    axs[2].set_title('Traffic Speed Over Time')
    axs[2].set_xlabel('Frame')
    axs[2].set_ylabel('Speed')
    axs[2].legend()
    axs[2].grid()
    
    # 调整布局
    plt.tight_layout()
    
    # 显示图形
    plt.show()
```

```python
for i in csv_list:
    lrm(i)
```
