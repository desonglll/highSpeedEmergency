```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to read CSV and plot the data trends for Flow, Density, and Speed against Frame
def plot_trends(csv_path):
    # Read CSV data
    data = pd.read_csv(csv_path)

    # Create subplots for Flow, Density, and Speed
    fig, axs = plt.subplots(3, 1, figsize=(8, 10))

    # Plot Flow vs Frame
    axs[0].plot(data['Frame'], data['Flow'], color='b', label='Flow')
    axs[0].set_title('Flow vs Frame')
    axs[0].set_xlabel('Frame')
    axs[0].set_ylabel('Flow')
    axs[0].grid(True)
    
    # Plot Density vs Frame
    axs[1].plot(data['Frame'], data['Density'], color='g', label='Density')
    axs[1].set_title('Density vs Frame')
    axs[1].set_xlabel('Frame')
    axs[1].set_ylabel('Density')
    axs[1].grid(True)
    
    # Plot Speed vs Frame
    axs[2].plot(data['Frame'], data['Speed'], color='r', label='Speed')
    axs[2].set_title('Speed vs Frame')
    axs[2].set_xlabel('Frame')
    axs[2].set_ylabel('Speed')
    axs[2].grid(True)

    # Adjust layout
    plt.tight_layout()

    # Show plots
    plt.show()
```

```python
plot_trends("./res/with-speed/32.31.250.103/20240501_20240501125647_20240501140806_125649.csv")
```
