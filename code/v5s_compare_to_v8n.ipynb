```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV 
csv_v5 = 'yolov5_runtime_stats.csv'  #  YOLOv5 
csv_v8 = 'yolov8_runtime_stats.csv'  #  YOLOv8 

# 
df_v5 = pd.read_csv(csv_v5)
df_v8 = pd.read_csv(csv_v8)

# 
video_names = df_v5['Video Name']
v5_processed_frames = df_v5['Processed Frames']
v8_processed_frames = df_v8['Processed Frames']

v5_total_time = df_v5['Total Time (s)']
v8_total_time = df_v8['Total Time (s)']

v5_process_fps = df_v5['Process FPS']
v8_process_fps = df_v8['Process FPS']

# 
index_labels = range(1, len(video_names) + 1)

# 
plt.figure(figsize=(12, 8))

# 
plt.subplot(3, 1, 1)
plt.plot(index_labels, v5_processed_frames, label='YOLOv5 Processed Frames', marker='o', color='b')
plt.plot(index_labels, v8_processed_frames, label='YOLOv8 Processed Frames', marker='o', color='g')
plt.title('Processed Frames Comparison')
plt.xlabel('Video Index')
plt.ylabel('Processed Frames')
plt.xticks(index_labels)  # x
plt.legend()

# 
plt.subplot(3, 1, 2)
plt.plot(index_labels, v5_total_time, label='YOLOv5 Total Time (s)', marker='o', color='b')
plt.plot(index_labels, v8_total_time, label='YOLOv8 Total Time (s)', marker='o', color='g')
plt.title('Total Time Comparison')
plt.xlabel('Video Index')
plt.ylabel('Total Time (s)')
plt.xticks(index_labels)  # x
plt.legend()

# 
plt.subplot(3, 1, 3)
plt.plot(index_labels, v5_process_fps, label='YOLOv5 Process FPS', marker='o', color='b')
plt.plot(index_labels, v8_process_fps, label='YOLOv8 Process FPS', marker='o', color='g')
plt.title('Process FPS Comparison')
plt.xlabel('Video Index')
plt.ylabel('Process FPS')
plt.xticks(index_labels)  # x
plt.legend()

# 
plt.tight_layout()
plt.show()
```
