```python
from tools.csv import csv_files

csv_paths = csv_files("./res/with-speed-combined")
print(csv_paths)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def normalize_and_plot(csv_path,ip, arg):
    # 读取CSV数据
    data = pd.read_csv(csv_path)
    
    # 归一化处理
    data['Flow'] = (data['Flow'] - data['Flow'].min()) / (data['Flow'].max() - data['Flow'].min())
    data['Density'] = (data['Density'] - data['Density'].min()) / (data['Density'].max() - data['Density'].min())
    data['Speed'] = (data['Speed'] - data['Speed'].min()) / (data['Speed'].max() - data['Speed'].min())

    # 计算时间（分钟）
    data['Minute'] = data['Frame'] / arg

    # 绘制图形
    plt.figure(figsize=(12, 6))
    plt.plot(data['Minute'], data['Flow'], label='Flow (Normalized)', color='blue')
    plt.plot(data['Minute'], data['Density'], label='Density (Normalized)', color='orange')
    plt.plot(data['Minute'], data['Speed'], label='Speed (Normalized)', color='green')
    plt.title(f'Normalized Flow, Density, and Speed Over Time ({ip})')
    plt.xlabel('Minute')
    plt.ylabel('Normalized Value')
    plt.legend()
    plt.grid()
    plt.show()

# 示例用法
# normalize_and_plot('your_data.csv')
```

```python
args = [160.0, 150.0, 200.0, 200.0]
i = 0
for csv in csv_paths:
    ip_addr = csv.split("/")[-2]
    print(ip_addr)
    normalize_and_plot(csv, ip_addr, args[i])
    i+=1
    pass
```
